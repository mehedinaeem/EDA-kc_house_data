# EDA Project Structure: King County House Data
This notebook demonstrates how to create a clean folder structure for an exploratory data analysis (EDA) project using the King County house sales dataset.

## 1. Create Main Project Folders
We will use Python's `os` and `pathlib` modules to create the main project directory if it does not exist.

```python
import os
from pathlib import Path

# Define main project directory
main_dir = Path('EDA (kc_house_data)')
main_dir.mkdir(exist_ok=True)

print(f"Created main directory: {main_dir.resolve()}")
```

## 2. Create Data Folder and Add CSV File
Create a `data` folder inside the main directory and add a placeholder for `kc_house_data.csv`.

```python
# Create data folder and placeholder CSV

data_dir = main_dir / 'data'
data_dir.mkdir(exist_ok=True)

csv_path = data_dir / 'kc_house_data.csv'
if not csv_path.exists():
    csv_path.write_text('id,date,price,bedrooms,bathrooms,sqft_living,sqft_lot,floors,waterfront,view,condition,grade,sqft_above,sqft_basement,yr_built,yr_renovated,zipcode,lat,long,sqft_living15,sqft_lot15\n')
    print(f"Created placeholder CSV: {csv_path}")
else:
    print(f"CSV already exists: {csv_path}")
```

## 3. Create Notebooks Folder and Add EDA Notebook
Create a `notebooks` folder and add a placeholder file `01_eda.ipynb`.

```python
# Create notebooks folder and placeholder notebook
notebooks_dir = main_dir / 'notebooks'
notebooks_dir.mkdir(exist_ok=True)

notebook_path = notebooks_dir / '01_eda.ipynb'
if not notebook_path.exists():
    notebook_path.write_text('')
    print(f"Created placeholder notebook: {notebook_path}")
else:
    print(f"Notebook already exists: {notebook_path}")
```

## 4. Create Scripts Folder and Add main.py
Create a `scripts` folder and add a placeholder file `main.py`.

```python
# Create scripts folder and placeholder main.py
scripts_dir = main_dir / 'scripts'
scripts_dir.mkdir(exist_ok=True)

main_py_path = scripts_dir / 'main.py'
if not main_py_path.exists():
    main_py_path.write_text('')
    print(f"Created placeholder script: {main_py_path}")
else:
    print(f"Script already exists: {main_py_path}")
```

## 5. Create Outputs and Figures Folders
Create an `outputs` folder and a nested `figures` folder inside it.

```python
# Create outputs and figures folders
outputs_dir = main_dir / 'outputs'
outputs_dir.mkdir(exist_ok=True)

figures_dir = outputs_dir / 'figures'
figures_dir.mkdir(exist_ok=True)

print(f"Created outputs directory: {outputs_dir}")
print(f"Created figures directory: {figures_dir}")
```

## 6. Create README.md File
Create a `README.md` file in the main project directory with a short project description.

```python
# Create README.md file
readme_path = main_dir / 'README.md'
if not readme_path.exists():
    readme_content = """# EDA (KC_house_data)

This project contains exploratory data analysis (EDA) for the King County House Sales dataset.

## Folder Structure

- data/: Dataset
- notebooks/: Jupyter notebooks
- scripts/: Python scripts
- outputs/figures/: Generated figures
"""
    readme_path.write_text(readme_content)
    print(f"Created README.md: {readme_path}")
else:
    print(f"README.md already exists: {readme_path}")
```
